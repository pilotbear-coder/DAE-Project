```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Step 1: Data Import and Formatting

```python
# Read the dataset into a DataFrame
df = pd.read_csv("athlete_events.csv") 
df
```

#### Step 2: Filter the DataFrame only to include the rows corresponding to medal winners in 2016 

##### a. Filter the DataFrame to include only records from the year 2016 

```python

# Filter the DataFrame to include only records from the year 2016 and store in temp_df
temp_df = df[df['Year']==2016]
temp_df
```

```python
# Print first 5 rows of the temporary dataframe
temp_df.head()
```

##### b. Find the number of null and non-null values in 'Medal Column'

```python


null_summary = temp_df['Medal'].isnull().value_counts()


null_summary
```

##### c. Remove rows where the 'Medal' column has missing values.

```python


temp_df = temp_df.dropna(subset=['Medal'])


temp_df
```

```python
temp_df.head()
```

#### Step 3: Find out the medals awarded in 2016 for each sport. 

```python

sport_medal_summary = temp_df[temp_df['Year'] == 2016].groupby('Sport')['Medal'].count()


sport_medal_summary
```

#### Step 4: List the top five sports based on the most significant number of medals awarded in 2016. 

```python

top_five_sports = sport_medal_summary.sort_values(ascending=False).head(5)


top_five_sports
```

#### Step 5: Filter the DataFrame one more time to include the records for the top five sports in 2016.

```python

top_five_sport_names = top_five_sports.index.tolist()
top5_df = temp_df[(temp_df['Year'] == 2016) & (temp_df['Sport'].isin(top_five_sport_names))]



top5_df
```

```python
top5_df.head()
```

#### Step 6: Generate a bar plot of record counts corresponding to the top five sports in 2016. 

```python

'''   
sport_counts = top5_df['Sport'].value_counts()

plt.figure(figsize=(10, 6))
sport_counts.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Top 5 Sports by Record Count in 2016 Olympics', fontsize=16, fontweight='bold')
plt.xlabel('Sports', fontsize=12)
plt.ylabel('Number of Records', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.show()
'''
##Matplotlib Option
fig, ax = plt.subplots(figsize=(10, 6))
sport_counts = top5_df['Sport'].value_counts()
ax.bar(sport_counts.index, sport_counts.values)

ax.set_title('Record Counts of Top 5 Sports in 2016 Olympics')
ax.set_xlabel('Sports')
ax.set_ylabel('Number of Records')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Step 7: Generate a histogram for the Age feature of all medal winners in the top five sports in 2016

```python


'''
plt.figure(figsize=(10, 6))
age_histogram = plt.hist(top5_df['Age'], bins=20, color='lightblue', edgecolor='black', alpha=0.7)

plt.title('Age Distribution of Athletes in Top 5 Sports (2016 Olympics)', fontsize=16, fontweight='bold')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)

plt.show()
age_histogram
'''

plt.figure(figsize=(10, 6))
age_counts = top5_df['Age'].value_counts()
plt.bar(age_counts.index, age_counts.values)

plt.title('Age Distribution of Athletes in Top 5 Sports (2016 Olympics)')
plt.xlabel('Age')
plt.ylabel('Number of Athletes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Step 8: Generate a bar plot indicating how many medals were won by each country's team in the top five sports in 2016, and sort them. Make sure to set the figure size to 15x5

```python

team_medal_counts = top5_df['NOC'].value_counts().head(10)

fig = plt.figure(figsize=(15, 5))
final_bar_plot = plt.bar(team_medal_counts.index, team_medal_counts.values, 
                        color='gold', edgecolor='black', alpha=0.8)

plt.title('Medals Won by Each Team in Top 5 Sports (2016 Olympics)', fontsize=16, fontweight='bold')
plt.xlabel('Team (NOC)', fontsize=12)
plt.ylabel('Number of Medals', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.show()
final_bar_plot
```
