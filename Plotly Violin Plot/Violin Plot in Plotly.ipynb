```python
import pandas as pd
import plotly.express as px
```

```python
diamonds = pd.read_csv("diamonds.csv")
diamonds
```

```python
fig = px.violin(diamonds,x='cut',y='price',title='Distribution of Diamonds Prices Across Different Cuts')
fig.show()
```

```python
fig = px.violin(diamonds,x='cut',y='price',box=True,title='Distribution of Diamonds Prices Across Different Cuts')
fig.show()
```

```python
fig = px.imshow(
    diamonds.corr(),
    color_continuous_scale="Inferno_r",
)

fig.show()
```
