```python
numeric_vars = ['Order_ID', 'Units_Sold', 'MRP', 'Discount_Applied', 'Revenue', 'Profit']


for var in numeric_vars:
    plt.figure(figsize=(8, 5))
    sns.kdeplot(data=df, x=var, fill=True, linewidth=2, color='skyblue')
    plt.title(f'KDE Plot for {var}', fontsize=14)
    plt.xlabel(var, fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
