```python
# Import necessary libraries.
import pandas as pd
import matplotlib.pyplot as plt

# Read data from 'gdp_csv.csv'.
df = pd.read_csv('gdp_csv.csv')

# Display the data.
df
```

```python
# Extract data for specific regions using the 'Country Name' column
df_eu = df.loc[df['Country Name'] == 'European Union']
df_na = df.loc[df['Country Name'] == 'North America']
df_sa = df.loc[df['Country Name'] == 'South Asia']
df_ea = df.loc[df['Country Name'] == 'East Asia & Pacific']

# Display the first few rows of each regional dataset
print(df_eu.head())
print(df_na.head())
print(df_sa.head())
print(df_ea.head())
```

```python
# Create a plot for GDP per year for each region
fig, ax = plt.subplots()
ax.plot(df_eu['Year'], df_eu['Value'], label='European Union GDP per Year')
ax.plot(df_na['Year'], df_na['Value'], label='North America GDP per Year')
ax.plot(df_sa['Year'], df_sa['Value'], label='South Asia GDP per Year')
ax.plot(df_ea['Year'], df_ea['Value'], label='East Asia & Pacific GDP per Year')

# Add a legend to the plot
ax.legend()

# Display the plot
plt.show()
```

### Plotting a Line Plot Logarithmically

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate an array of numbers (start, end, number of samples)
x = np.linspace(0, 5, 100)

# Calculate y-values as the exponential of x
y = np.exp(x)

# Create a line plot of x and y
plt.plot(x, y)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate an array of x-values from 0 to 5 with 10 samples.
x = np.linspace(0, 5, 100)

# Calculate y-values as the exponential of x.
y = np.exp(x)

# Set the scale of the Y-axis to logarithmic.
plt.yscale('log')

# Create a line plot of x and y with a logarithmic Y-axis scale.
plt.plot(x, y)

# Display the plot.
plt.show()
```

### Customizing Line Plots 

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the GDP dataset from a CSV file
df = pd.read_csv('gdp_csv.csv')

# Extract data for specific regions using the 'Country Name' column
df_eu = df.loc[df['Country Name'] == 'European Union']
df_na = df.loc[df['Country Name'] == 'North America']
df_sa = df.loc[df['Country Name'] == 'South Asia']
df_ea = df.loc[df['Country Name'] == 'East Asia & Pacific']

# Create a new figure and axis for the plot
fig, ax = plt.subplots()

# Plot the GDP per year for each region and label the lines with customizations
ax.plot(df_eu['Year'], df_eu['Value'], label='European Union GDP per Year', linestyle='dotted', color='b')
ax.plot(df_na['Year'], df_na['Value'], label='North America GDP per Year', linestyle='solid', color='g')
ax.plot(df_sa['Year'], df_sa['Value'], label='South Asia GDP per Year', linestyle='dashed', color='r')
ax.plot(df_ea['Year'], df_ea['Value'], label='East Asia & Pacific GDP per Year', linestyle='dashdot', color='m')

# Change the Y-axis scale to logarithmic for better visualization
plt.yscale('log')

# Add a legend to the plot
ax.legend()
plt.title('GPD by Region')

# Display the plot
plt.show()
```
