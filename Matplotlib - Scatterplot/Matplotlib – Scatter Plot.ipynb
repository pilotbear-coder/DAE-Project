```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv('AmesHousing.csv')

# Display the dataset
df.head()
```

### Plotting a Scatter Plot 

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot using 'Gr Liv Area' on the x-axis and 'SalePrice' on the y-axis
ax.scatter(x=df['Gr Liv Area'], y=df['SalePrice'])

# Set the title for the scatter plot
plt.title("Relationship between Living Area Above Ground and House Price")

# Set the label for the x-axis
plt.xlabel("Living Area Above Ground")

# Set the label for the y-axis
plt.ylabel("House Price")

# Display the scatter plot
plt.show()
```

```python
# Create a figure with 2 side-by-side subplots, each with a size of 10 units by 6 units
fig, ax = plt.subplots(2, figsize=(10, 6))

# Top subplot: Scatter plot showing the relationship between 'Living Area Above Ground' and 'House Price'
ax[0].scatter(x=df['Gr Liv Area'], y=df['SalePrice'])
ax[0].set_xlabel("Living Area Above Ground")  # Label for the x-axis
ax[0].set_ylabel("House Price")  # Label for the y-axis
ax[0].set_title("Living Area vs. House Price")  # Title for the top subplot

# Bottom subplot: Scatter plot showing the relationship between 'Overall Quality' and 'House Price'
ax[1].scatter(x=df['Overall Qual'], y=df['SalePrice'])
ax[1].set_xlabel("Overall Quality")  # Label for the x-axis
ax[1].set_ylabel("House Price")  # Label for the y-axis
ax[1].set_title("Overall Quality vs. House Price")  # Title for the bottom subplot

plt.tight_layout()  # Adjust layout for better spacing

# Display the subplots
plt.show()
```

### Changing the Marker type 

```python
# Create a figure with 2 side-by-side subplots, each with a size of 10 units by 6 units
fig, ax = plt.subplots(2, figsize=(10, 6))

# Top subplot: Scatter plot showing the relationship between 'Living Area Above Ground' and 'House Price'
ax[0].scatter(x=df['Gr Liv Area'], y=df['SalePrice'])
ax[0].set_xlabel("Living Area Above Ground")  # Label for the x-axis
ax[0].set_ylabel("House Price")  # Label for the y-axis
ax[0].set_title("Living Area vs. House Price")  # Title for the top subplot

# Bottom subplot: Scatter plot showing the relationship between 'Overall Quality' and 'House Price' with a marker style of '_'
ax[1].scatter(x=df['Overall Qual'], y=df['SalePrice'], marker='_')
ax[1].set_xlabel("Overall Quality")  # Label for the x-axis
ax[1].set_ylabel("House Price")  # Label for the y-axis
ax[1].set_title("Overall Quality vs. House Price")  # Title for the bottom subplot

plt.show()
```

### Customizing Scatter Plot in Matplotlib

```python
# Create a figure with a single subplot and a specified size of 10 units by 6 units
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot showing the relationship between 'Living Area Above Ground' and 'House Price'
ax.scatter(x=df['Gr Liv Area'], y=df['SalePrice'],
           color='orange',         # Set the color of the markers to blue
           edgecolor='white',    # Set the edge color of the markers to white
           linewidths=0.1,       # Set the width of marker edges to 0.1 units
           alpha=0.7)            # Set the transparency of markers to 0.7 (70%)

# Set the label for the x-axis
plt.xlabel("Living Area Above Ground")

# Set the label for the y-axis
plt.ylabel("House Price")

# Set the title for the scatter plot
plt.title("Relationship between Living Area Above Ground and House Price")

# Display the scatter plot
plt.show()
```

```python
# Create a figure with a single subplot and a specified size of 10 units by 6 units
fig, ax = plt.subplots(figsize=(10, 6))

# Define a dictionary of colors associated with different sale conditions
colors = {'Normal': '#db5f57',
          'Abnorml': '#d3db57',
          'Family': '#57db5f',
          'Partial': '#57d3db',
          'Alloca': '#5f57db',
          'AdjLand': '#db57d3'}

# Scatter plot with color-coded markers based on the 'Sale Condition'
ax.scatter(x=df['Overall Qual'], y=df['SalePrice'],
           marker='_',  # Set the marker style to '_'
           color=df['Sale Condition'].apply(lambda x: colors[x]))  # Color mapping based on sale condition

# Set the label for the x-axis
plt.xlabel("Overall Quality")

# Set the label for the y-axis
plt.ylabel("House Price")

# Set the title for the scatter plot
plt.title("Relationship between Overall Quality and House Price with Sale Condition Variation")

# Display the scatter plot
plt.show()
```

### Bubble Plot

```python
# Load the dataset
df = pd.read_csv('worldHappiness2019.csv')

# Disply df
df.head()
```

```python
# Extract the 'Score' column from the DataFrame and convert it to a NumPy array
size = df['Score'].to_numpy()

# Use list comprehension to create a list of sizes for markers based on the cube of each value in 'size'
s = [3 * s**3 for s in size]

# Create a figure with a single subplot and a specified size of 10 units by 6 units
fig, ax = plt.subplots(figsize=(10, 6))

# Create scatterplot
ax.scatter(x=df['GDP per capita'], y=df['Generosity'], s=s, alpha=0.5)

# Set the label for the x-axis
plt.xlabel("GDP per Capita")

# Set the label for the y-axis
plt.ylabel("Generosity Score")

# Set the title for the bubble plot
plt.title("Relationship between GDP per Capita, Generosity, and Country Happiness")

# Display the scatter plot
plt.show()
```
