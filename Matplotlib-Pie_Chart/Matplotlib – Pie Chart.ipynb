```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Plotting a Pie Chart

```python
# Data representing the sizes of different categories
x = [15, 25, 25, 30, 5]

# Create a pie chart using Matplotlib
fig, ax = plt.subplots()
ax.pie(x)

# Display the pie chart
plt.show()
```

### Adding Labels to the pie chart

```python
# Data representing the sizes of different categories
x = [15, 25, 25, 30, 5]

# Labels corresponding to each category
labels = ['Very Likely', 'Likely', 'Unsure', 'Unlikely', 'Very Unlikely']

# Create a pie chart with labels using Matplotlib
fig, ax = plt.subplots()
ax.pie(x, labels=labels)

# Set a title for the pie chart
ax.set_title('Survey Response')

# Display the pie chart
plt.show()
```

### Changing Pie Chart Colors 

```python
# Data representing the sizes of different categories
x = [15, 25, 25, 30, 5]

# Labels corresponding to each category
labels = ['Very Likely', 'Likely', 'Unsure', 'Unlikely', 'Very Unlikely']

# An array of colors for each category
colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red']

# Create a pie chart with labels and custom colors using Matplotlib
fig, ax = plt.subplots()
ax.pie(x, labels=labels, colors=colors)

# Set a title for the pie chart
ax.set_title('Survey Response')

# Display the pie chart
plt.show()
```

### Showing Percentages on Slices 

```python
# Define the data for the pie chart
x = [15, 25, 25, 30, 5]

# Define labels for each segment
labels = ['Very Likely', 'Likely', 'Unsure', 'Unlikely', 'Very Unlikely']

# Define custom colors for each segment
colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red']

# Create a subplot and plot the pie chart
fig, ax = plt.subplots()
ax.pie(x, labels=labels, colors=colors, autopct='%.0f%%')

# Set the title of the pie chart
ax.set_title('Survey Response')

# Display the pie chart
plt.show()
```

### Explode Slices 

```python
# Define a list of values representing survey responses
x = [15, 25, 25, 30, 5]

# Define labels for each segment
labels = ['Very Likely', 'Likely', 'Unsure', 'Unlikely', 'Very Unlikely']

# Define custom colors for each segment
colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red']

# Define the extent to which the slices of the pie should be separated (explode the 'Very Unlikely' category)
explode = [0, 0, 0, 0, 0.2]

# Create a subplot and plot the pie chart
fig, ax = plt.subplots()
ax.pie(x, labels=labels, colors=colors, autopct='%.0f%%', explode=explode)

# Set the title of the pie chart
ax.set_title('Survey response')

# Display the pie chart
plt.show()
```

### Adding Shadow and Rotating Pie Chart 

```python
# Define a list of values representing survey responses
x = [15, 25, 25, 30, 5]

# Define labels for each segment
labels = ['Very Likely', 'Likely', 'Unsure', 'Unlikely', 'Very Unlikely']

# Define custom colors for each segment
colors = ['tab:blue', 'tab:cyan', 'tab:gray', 'tab:orange', 'tab:red']

# Define the extent to which the slices of the pie should be separated (explode the 'Very Unlikely' category)
explode = [0, 0, 0, 0, 0.2]

# Create a subplot and plot the pie chart
fig, ax = plt.subplots()
ax.pie(x, labels=labels, colors=colors, autopct='%.0f%%', explode=explode, shadow=True, startangle=180)

# Set the title of the pie chart
ax.set_title('Survey response')

# Display the pie chart
plt.show()
```

### Using Dataset to Create Pie Charts 

```python
# Read the World Happiness Report 2021 dataset into a DataFrame
df = pd.read_csv('world-happiness-report-2021.csv')

# Group the DataFrame by "Regional indicator" and calculate the mean values for each group
df_mean = df.groupby("Regional indicator").mean()

# Display the mean perceptions of corruption for different regions
df_mean['Perceptions of corruption']
```

```python
# Create a figure and axis 
fig, ax = plt.subplots()

# Define a list of custom colors
colors = ['#db5f57', '#d3db57', '#57db5f', '#57d3db', '#5f57db', '#db57d3']

# Plot a pie chart using data from the 'Perceptions of corruption' column in the DataFrame 'df_mean'
ax.pie(df_mean['Perceptions of corruption'], labels=df_mean.index, colors=colors, autopct='%.2f%%')

#Set the Title
plt.title("Perceptions of Corruption by Region", fontsize=16)

# Display the pie chart
plt.show()
```
