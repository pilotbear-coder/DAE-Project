```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('AmesHousing.csv')

#Display the data
df.head()
```

```python
# Calculate Pearson correlation coefficients for selected columns
pearson_corr = df.corr(method='pearson')[['SalePrice', 'Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Garage Area']]

# Create a subplot for a matrix plot
fig, ax = plt.subplots(figsize=(12, 6))

# Display the matrix plot using the matshow function
ax.matshow(pearson_corr)

# Set y-axis ticks with corresponding column names
plt.yticks(range(0, len(pearson_corr)), pearson_corr.index, fontsize=6)

# Set x-axis ticks with corresponding column names, rotate labels for better readability
plt.xticks(range(0, len(pearson_corr.columns)), pearson_corr.columns, fontsize=6, rotation=90)

# Display the matrix plot
plt.show()
```

```python
# Calculate Pearson correlation coefficients for 'SalePrice' and sort them in descending order
pearson_corr = df.corr(method='pearson')[['SalePrice']].sort_values(by='SalePrice', ascending=False)

# Create a subplot for a matrix plot
fig, ax = plt.subplots(figsize=(12, 6))

# Display the matrix plot using the matshow function
ax.matshow(pearson_corr, aspect=0.05)

# Set y-axis ticks with corresponding variable names
plt.yticks(range(0, len(pearson_corr)), pearson_corr.index, fontsize=6)

# Set x-axis ticks with corresponding variable names, rotate labels for better readability
plt.xticks(range(0, len(pearson_corr.columns)), pearson_corr.columns, fontsize=6, rotation=90)

# Add correlation values as text annotations
for y in range(pearson_corr.shape[0]):
    plt.text(0, y, '%.1f' % pearson_corr.iloc[y],
             horizontalalignment='center',
             verticalalignment='center',
             fontsize=6
             )

plt.show()
```

```python
# Calculate Pearson correlation coefficients for all features
pearson_corr = df.corr(method='pearson')

# Create a subplot for a matrix plot
fig, ax = plt.subplots(figsize=(12, 6))

# Display the matrix plot using the matshow function
ax.matshow(pearson_corr)

# Set y-axis ticks with corresponding variable names
plt.yticks(range(0, len(pearson_corr)), pearson_corr.index, fontsize=6)

# Set x-axis ticks with corresponding variable names, rotate labels for better readability
plt.xticks(range(0, len(pearson_corr.columns)), pearson_corr.columns, fontsize=6, rotation=90)

# Add correlation values as text annotations
for y in range(pearson_corr.shape[0]):
    for x in range(pearson_corr.shape[1]):
        plt.text(x, y, '%.1f' % pearson_corr.iloc[x, y],
                 horizontalalignment='center',
                 verticalalignment='center',
                 fontsize=6
                 )

plt.show()
```
