```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('world-happiness-report-2021.csv')

# Display the df
df
```

```python
# Extract columns 'Regional indicator' and 'Freedom to make life choices' from the DataFrame
selected_columns = df[['Regional indicator', 'Freedom to make life choices']]

# Display the resulting DataFrame with the selected columns
selected_columns
```

```python
# Create a figure and axis with a specified size
fig, ax = plt.subplots(figsize=(12, 6))

# Plot a bar chart using 'Regional indicator' on the x-axis and 'Freedom to make life choices' on the y-axis
ax.bar(df['Regional indicator'], df['Freedom to make life choices'], label='Freedom score')

# Add a legend to the plot
ax.legend()

# Rotate and wrap x-axis labels for better readability
plt.xticks(rotation=10, wrap=True)

# Display the plot
plt.show()
```

```python
# Calculate the mean and median values for the 'Freedom to make life choices' feature
mean = df['Freedom to make life choices'].mean()
median = df['Freedom to make life choices'].median()

# Create a figure and axisa
fig, ax = plt.subplots(figsize=(12, 6))

# Plot a bar chart using 'Regional indicator' on the x-axis and 'Freedom to make life choices' on the y-axis
ax.bar(df['Regional indicator'], df['Freedom to make life choices'], label='Freedom score')

# Add horizontal lines for mean and median values
ax.axhline(mean, 0, 1, color='red', label='Mean Value for Dataset')
ax.axhline(median, 0, 1, color='green', label='Median Value for Dataset')

# Add a legend to the plot
ax.legend()

# Rotate and wrap x-axis labels 
plt.xticks(rotation=10, wrap=True)

# Display the plot
plt.show()
```

```python
# Group the DataFrame by "Regional indicator" and calculate the mean values for each group
df_means = df.groupby("Regional indicator").mean()

# Extract the mean values for the 'Freedom to make life choices' feature from the grouped DataFrame
freedom_means = df_means['Freedom to make life choices']

freedom_means
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(18, 6))

# Plot a bar chart using the index (regions) and mean values for 'Freedom to make life choices'
ax.bar(df_means.index, df_means['Freedom to make life choices'], label='Freedom score')

# Add horizontal lines for mean and median values of the original dataset
ax.axhline(mean, 0, 1, color='red', label='Mean')

# Add a horizontal line for the median value of the original dataset
ax.axhline(median, 0, 1, color='green', label='Median')

# Add a legend to the plot
ax.legend()

# Rotate and wrap x-axis labels for better readability
plt.xticks(rotation=25, wrap=True)

# Display the plot
plt.show()
```

### Bar Plot with Error Bars in Matplotlib 

```python
# Group the DataFrame by "Regional indicator" and calculate the standard deviation values for each group
df_stds = df.groupby("Regional indicator").std()

# Extract the standard deviation values for the 'Freedom to make life choices' feature from the grouped DataFrame
freedom_std_devs = df_stds['Freedom to make life choices']

freedom_std_devs
```

```python
# Create a figure and axis 
fig, ax = plt.subplots(figsize=(18, 6))

# Plot a bar chart using the index (regions), mean values, and error bars representing standard deviation
ax.bar(df_means.index, df_means['Freedom to make life choices'], label='Freedom score', yerr=df_stds['Freedom to make life choices'])

# Add horizontal lines for mean and median values of the original dataset
ax.axhline(mean, 0, 1, color='red', label='Mean')

# Add a horizontal line for the median value of the original dataset
ax.axhline(median, 0, 1, color='green', label='Median')

# Add a legend to the plot
ax.legend()

# Rotate and wrap x-axis labels for better readability
plt.xticks(rotation=10, wrap=True)

# Display the plot
plt.show()
```

### Changing Bar Plot Colors 

```python
# Create a figure and axis with a specified size
fig, ax = plt.subplots(figsize=(18, 6))

# Changing the color of bars
colors = ['#db5f57', '#d3db57', '#57db5f', '#57d3db', '#5f57db', '#db57d3']
ax.bar(df_means.index, df_means['Freedom to make life choices'], label='Freedom score', yerr=df_stds['Freedom to make life choices'], color=colors)

# Add horizontal lines for mean and median values of the original dataset
ax.axhline(mean, 0, 1, color='red', label='Mean')

# Add a horizontal line for the median value of the original dataset
ax.axhline(median, 0, 1, color='green', label='Median')

# Add a legend to the plot
ax.legend()

# Rotate and wrap x-axis labels for better readability
plt.xticks(rotation=10, wrap=True)

# Display the plot
plt.show()
```

### Plotting Horizontal Bar Plots 

```python
# Create a horizontal bar plot
fig, ax = plt.subplots(figsize=(18, 6))

# Changing the color of horizontal bars
colors = ['#db5f57', '#d3db57', '#57db5f', '#57d3db', '#5f57db', '#db57d3']
ax.barh(df_means.index, df_means['Freedom to make life choices'], label='Freedom score', xerr=df_stds['Freedom to make life choices'], color=colors)

# Add vertical lines for mean and median values of the original dataset
ax.axvline(mean, 0, 1, color='red', label='Mean')

# Add a vertical line for the median value of the original dataset
ax.axvline(median, 0, 1, color='green', label='Median')

# Add a legend to the plot
ax.legend()

# Rotate and wrap y-axis labels for better readability
plt.xticks(rotation=10, wrap=True)

# Display the plot
plt.show()
```
